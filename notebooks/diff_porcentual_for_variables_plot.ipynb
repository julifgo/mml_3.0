```python
#PLOT HEATMAP!!
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
def plot_heatmap(values,x_labels, y_labels, title, include_values=False, value_fmt='.3f'):
    plt.figure(figsize=(30,3))
    cmap = LinearSegmentedColormap.from_list('RedGreenRed', ['red','orange', 'white','orange', 'red'])
    ax = sns.heatmap(values.T, cmap='RdYlBu', xticklabels=y_labels,yticklabels=x_labels, linewidth = 0.05, annot=include_values, fmt=value_fmt, vmin=-25, vmax=25)
    # ax.invert_xaxis()
    plt.title(title)
    plt.show()
```

```python
plot_heatmap(df[['POP','EXLIFE']],["POP","EXLIFE"], df['Year'], f'Heatmap with porcentual differences between measured data and MML standard simulation outputs for Latin America', include_values=True)
```
